```python
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
import matplotlib as mpl

mpl.style.use('ggplot')
mpl.rcParams['font.family'] = ['Family1', 'Serif', 'cursive']
mpl.rcParams['font.size'] = 18
mpl.rcParams['figure.max_open_warning'] =30
```

```python
caminho = Path('dados/radar/')
arquivos = [i for i in caminho.glob('*.txt')]
ucolunas = [0, 1, 2, 3, 4, 5, 6, 7]
cabe = ['tempo', '81', '84', '87', '90', '93', '96', '99']


for i in arquivos:
    df = pd.read_csv(i, sep='\s+',skiprows=[0], usecols=ucolunas, names=cabe)
    x = df['tempo']
    y = range(81, 100, 3)
    z = df['81'].values,df['84'].values, df['87'].values,df['90'].values,df['93'].values,df['96'].values, df['99'].values

    fig = plt.figure(figsize=(14, 6))
    cs = plt.contour(x, y, z, cmap='viridis')
    im = plt.contourf(cs)
    plt.title('Teste de contorno')
    plt.ylabel('Km')
    plt.xlabel('Tempo')

    plt.colorbar(label='m/s')
    plt.savefig(f'{str(i)[:-4]}.jpeg')
  
    plt.show()
    
```
